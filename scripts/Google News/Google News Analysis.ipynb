# Topic modeling terhadap pemberitaan RUU Penyiaran dari Mei 2024 hingga Mei 2025

```python
# Pre-processing for topic modeling

# Setting-up and importing necessary libraries

import re
from nltk.corpus import stopwords
import nltk

# Downloading Indonesian stopwords
try:
    stopwords.words('indonesian')
except LookupError:
    print("Indonesian stopwords not found. Downloading now...")
    nltk.download('stopwords')
    print("Download complete.")

# Combine title and description for a richer text source
final_filtered_df['text_to_model'] = final_filtered_df['title'] + ' ' + final_filtered_df['description'].astype(str)

# Getting the stopwords
indonesian_stopwords = stopwords.words('indonesian')
custom_stopwords = ['ruu', 'penyiaran', 'dpr', 'ri', 'uud', 'indonesia', 'jakarta', 'mengatakan', 'menyebutkan', 'ujar']
all_stopwords = indonesian_stopwords + custom_stopwords
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove numbers and punctuation
    words = text.split()
    # Remove stopwords
    words = [word for word in words if word not in all_stopwords]
    return ' '.join(words)

# Apply the preprocessing function to our text column
final_filtered_df['cleaned_text'] = final_filtered_df['text_to_model'].apply(preprocess_text)
topic_docs = final_filtered_df['cleaned_text'].tolist()

print("Preprocessing complete. Sample cleaned text:")
print(topic_docs[0])
```

```python
# Run the BERTopic

# Importing necessary libraries
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

# Multilingual model that understands Indonesian well.
embedding_model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Create the BERTopic model.
# We set `min_topic_size` to 3 to avoid creating tiny, insignificant topics.
# `nr_topics='auto'` will automatically find the best number of topics.
topic_model = BERTopic(
    embedding_model = embedding_model,
    min_topic_size = 3,
    nr_topics = 'auto',
    verbose = True
)

# Fit the model to our documents
topics, probabilities = topic_model.fit_transform(topic_docs)
```

```python
# Dataframe of most topics

topic_info = topic_model.get_topic_info()
print("\nTop Topics Found:")
print(topic_info)
```

```python
# Top 10 most topics

print("\nMost Frequent Topics:")
print(topic_model.get_topic_freq().head(10))
```

```python
# Visualizing topic modeling

barchart_fig = topic_model.visualize_barchart(top_n_topics = 13)
barchart_fig.show()
```

```python
# Intertopics relationship

intertopic_map_fig = topic_model.visualize_topics()
intertopic_map_fig.show()
```

# Framing analysis of pemberitaan RUU Penyiaran

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Defining the frames
frames = {
    "Press Freedom & Journalism": [
        'kebebasan pers', 'jurnalisme investigasi', 'dewan pers', 'wartawan',
        'jurnalis', 'liputan', 'independensi', 'kriminalisasi pers'
    ],
    "Digital Economy & Creators": [
        'kreator konten', 'platform digital', 'media sosial', 'influencer',
        'youtube', 'tiktok', 'iklan digital', 'ekonomi digital', 'svod', 'ott'
    ],
    "Political & Legal Process": [
        'komisi i', 'pemerintah', 'draf', 'pasal', 'revisi', 'baleg',
        'undang-undang', 'legislasi'
    ]
}

# Combine title and description for a comprehensive text search field.
final_filtered_df['full_text'] = final_filtered_df['title'].astype(str) + ' ' + final_filtered_df['description'].astype(str)
```

```python
# Categorize Each Article into a Frame
def assign_frame(text):
    text = str(text).lower()
    for frame, keywords in frames.items():
        if any(keyword in text for keyword in keywords):
            return frame # Return the first frame that matches
    return 'Other' # Assign 'Other' if no keywords match

final_filtered_df['frame'] = final_filtered_df['full_text'].apply(assign_frame)
```

```python
# Prepare the Data for Plotting ---
# We want to see the distribution of frames for the top news sources.
top_n_sources = final_filtered_df['source'].value_counts().nlargest(20).index

# Filter the DataFrame to only include these top sources.
df_top_sources = final_filtered_df[final_filtered_df['source'].isin(top_n_sources)]

# Create a summary table counting frames per source.
summary_table = pd.crosstab(df_top_sources['source'], df_top_sources['frame'])
summary_table_percent = summary_table.div(summary_table.sum(axis=1), axis=0) * 100

print(summary_table_percent)
```

```python
# Visualizing

plt.style.use('seaborn-v0_8-whitegrid')
ax = summary_table_percent.plot(
    kind='barh', # Horizontal bar chart
    stacked=True,
    figsize=(14, 10),
    colormap='viridis' # A nice color scheme
)

ax.set_title('Framing of "RUU Penyiaran" by Top 20 Media Outlets', fontsize=20, pad=20)
ax.set_xlabel('Percentage of Articles (%)', fontsize=12)
ax.set_ylabel('News Source', fontsize=12)
ax.legend(title='Frame', bbox_to_anchor=(1.02, 1), loc='upper left') # Move legend outside
ax.invert_yaxis() # Show the source with the most articles at the top
ax.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```

# Analisis tren kata kunci (keywords) pemberitaan RUU Penyiaran

```python
# Setting up the date

date_column_source = None
if 'published_date' in final_filtered_df.columns:
    date_column_source = 'published_date'
elif 'date_string' in final_filtered_df.columns:
    print("Warning: 'published_date' column not found. Using 'date_string' as a fallback.")
    date_column_source = 'date_string'
else:
    print("Error: Could not find a date column named 'published_date' or 'date_string'.")
    # In a real notebook, you would stop here. For this script, we'll exit.
    exit()

final_filtered_df['published_date'] = pd.to_datetime(final_filtered_df[date_column_source], errors='coerce')
final_filtered_df.dropna(subset=['published_date'], inplace=True)

# Combine title and description
final_filtered_df['full_text'] = final_filtered_df['title'].astype(str) + ' ' + final_filtered_df['description'].astype(str)
```

```python
# Create an empty DataFrame to hold our trend data
trends_df = pd.DataFrame()

# Loop through each frame to calculate its daily trend
for frame_name, keywords in frames.items():
    # Create a regex pattern to search for any of the keywords in the frame
    keyword_pattern = '|'.join(keywords)
    
    # Create a boolean series: True if an article contains any of the keywords
    contains_keywords = final_filtered_df['full_text'].str.contains(keyword_pattern, case=False, na=False)
    
    # Create a temporary DataFrame with only the relevant articles for this frame
    frame_df = final_filtered_df[contains_keywords]
    
    # Resample by Day ('D') and count the articles.
    # You could also use 'W' for weekly trends to smooth the lines.
    daily_counts = frame_df.set_index('published_date').resample('D').size()
    
    # Add the result as a new column in our trends_df
    trends_df[frame_name] = daily_counts

# Fill any missing days with 0 articles
trends_df = trends_df.fillna(0)

print("\n--- Preview of Trend Data ---")
print(trends_df.head())
```

```python
# Visualize the Trends

plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(18, 9))

# Plot each column from our trends_df as a separate line
trends_df.plot(
    ax=ax,
    linewidth=2.5,
    alpha=0.8
)

ax.set_title('Evolution of Media Frames in "RUU Penyiaran" Coverage', fontsize=20, pad=20)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Number of Articles Published', fontsize=12)
ax.legend(title='Debate Frame', fontsize='medium', loc='upper left')
ax.spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```
